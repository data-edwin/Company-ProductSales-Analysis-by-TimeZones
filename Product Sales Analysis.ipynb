# Product Sales Analysis by Time Zone

### Import Libraries

```python
import pandas as pd
import pytz
from datetime import datetime
from matplotlib import pyplot as plt
```

### Load and Explore Data

```python
data = pd.read_csv('Sales-products-tz-mod.csv', index_col = 'SaleID')
sales_data = data.copy()
sales_data
```

```python
# Let's see the date, time and time-zone values

sales_data[['DateOfSale', "TimeOfSale", "TimeZone"]].head()
```

### Data Preparation

```python
# Let's combine the dateofsale and timeofsale columns concatenated with a separating white space

date_and_time = sales_data['DateOfSale'] + " " + sales_data['TimeOfSale']
date_and_time 
```

```python
# Now, we'll convert this to datetime object and store in a new column MOS (moment of sale)

sales_data['MOS'] = pd.to_datetime(date_and_time)
sales_data.head()
```

```python
# Let's remove dateofsale and timeofsale columns since we have that information in MOS

sales_data = sales_data.drop(['DateOfSale','TimeOfSale'], axis = 1)
sales_data.head()
```

```python
# Let's use the pytz module to view all avilable olson time zones in python
# (run the code twice for better arrangement)

pytz.all_timezones
```

```python
# Let's see the time zones in our dataset

sales_data['TimeZone'].unique()
```

```python
# We verify if the time zones in our database can be found in the pytz module database

for i in sales_data['TimeZone'].unique():
    if i in (pytz.all_timezones):
        print (True)
    else:
        print (False)
```

```python
# Let's confirm which time zone returned 'False'

sales_data['TimeZone'].unique()
```

```python
# SGT, signifying Singapore returned 'false'
# The pytz module list represents this as 'Singapore', so let's substitute the value in our dataset

sales_data.loc[:, 'TimeZone'] = sales_data.loc[:, 'TimeZone'].replace({'SGT': 'Singapore'})
```

```python
# Let's see the unique values again

sales_data['TimeZone'].unique()
```

```python
# Confirming correspondence

for i in sales_data['TimeZone'].unique():
    if i in (pytz.all_timezones):
        print (True)
    else:
        print (False)
```

```python
# Further check if 'Singapore' is well represented in the dataset

sales_data[sales_data['TimeZone'] == 'Singapore']
```

```python
# View entire data

pd.options.display.max_rows = None
pd.options.display.max_columns = None
sales_data
```

### Data Manipulation

```python
# We want to estimate the offset of UTC from MOS
# Let's consider the first row

sales_data['MOS'].iloc[0]
```

```python
# Next, we obtain the timezone indication associated with the given sale of the timezone column.
# This will also transform the object from naive to timezone aware.

sales_data['MOS'].iloc[0].tz_localize(sales_data['TimeZone'].iloc[0])

# This shows the timestamp data aware of the offset from UTC and the corresponding timezone.
```

```python
# Create timezone aware records for all records.
# This is to obtain the relevant offset values and time zones for each record.
# The list will hold all timestamp values for all our data

list_mos_timestamps = []
list_mos_timestamps
```

```python
# Let's append to this list

for i in range(len(sales_data)):
    list_mos_timestamps.append(pd.to_datetime(sales_data['MOS'].iloc[i].tz_localize(sales_data['TimeZone'].iloc[i], ambiguous='raise', nonexistent='raise')))

# The raise value allows us to obtain a message indicating an exception if a certain condition has not been met.
# In this case, no exception was met
```

```python
# Let's display a few values from the list

list_mos_timestamps[:10]
```

```python
# Let's check the entire list to be sure

list_mos_timestamps
```

- Obtaining different information from the timestamp data

```python
# Obtaining only the timestamp info
list_mos_timestamps[0].strftime('%Y-%m-%d %H:%M:%S')
```

```python
# Obtaining UTC offset
list_mos_timestamps[0].strftime('%z')

# This says 5 hours behind UTC
```

```python
# Let's represent the above result in numeric
float(list_mos_timestamps[0].strftime('%z'))/100
```

```python
# Obtaining the date value
list_mos_timestamps[0].date()
```

```python
# Obtaining the time value
list_mos_timestamps[0].time()
```

```python
# Obtaining UTC offset in a different method
list_mos_timestamps[0].utcoffset()

# This says -1 days, + 68400 seconds
```

```python
# Let's convert that to hours
list_mos_timestamps[0].utcoffset().total_seconds() / 3600

# We get -5 hours just as obtained previously
```

```python
# We create a new column to store UTC offset in our sales data

sales_data['OffsetUTC'] = [list_mos_timestamps[i].utcoffset().total_seconds() / 3600 for i in range(len(sales_data))]
sales_data.head()
```

```python
# Let's re-order the columns

column_names_reordered = ['RetailerCountry', 'MOS', 'TimeZone', 'OffsetUTC', 'RetailerType', 'Product', 'Sales Revenue ($)']
sales_data = sales_data[column_names_reordered]
sales_data.head()
```

```python
# Let's see the offset values associated with the different time zones

sales_data.groupby(['TimeZone', 'OffsetUTC']).count()

# This shows that some time zones are associated with more than one offset values
# This difference is a result of the DST specifications related to the various regions and moments of sale
```

```python
# We sort UTC offset for CET timezone
sales_data[sales_data['TimeZone']=='CET'].sort_values('OffsetUTC')
```

### Data Analysis

```python
# We want to use MOS to order all the records chronologically
sales_data.sort_values(['MOS'])

# This will only order them according to the time and date in different time zones.
```

```python
# We want to reconcile moments of sale with their UTC-equivalent representations
# First, we need to convert all the values to UTC
# Next, we store these values in a different column
# Then we arrange these values chronologically

sales_data['MOS-UTC'] = [pd.to_datetime(sales_data['MOS'].iloc[i]).tz_localize(sales_data['TimeZone'].iloc[i]).tz_convert('UTC') 
                        for i in range(len(sales_data))]
```

```python
sales_data.head()
```

```python
# Re-order the columns

column_names_reordered = ['RetailerCountry', 'MOS-UTC', 'MOS', 'TimeZone', 'OffsetUTC', 'RetailerType', 'Product', 'Sales Revenue ($)']
sales_data = sales_data[column_names_reordered]
sales_data.head()
```

```python
# Now, let's order by these UTC moments of sale

sales_data.sort_values(['MOS-UTC'])
```

```python
# Next, we want to group the records by the values in the time zone column and see what's the average offset from UTC for that group

sales_data.groupby('TimeZone').mean()

# CET and EET return floats because different countries in same time zones have different offset from UTC due to DST regulations.
```

```python
# Count how many sales have occured in each time zone
sales_data.groupby('TimeZone').count()
```

```python
# This shows that there are no discrepancies, hence no missing data
# EET and Singapore has same number of sales but the sales revenue in EET is about 22 times higher than in Singapore, thus there's need to inquire about the data quality.
```

```python
# Groupby time zone and offset UTC
sales_data.groupby(['TimeZone', 'OffsetUTC']).count()
```

```python
# We want to obtain all sales grouped by time zones from the one with the highest number of sales to the one with the lowest

sales_data.groupby('TimeZone').count().sort_values('Product', ascending = False)
```

### Data Visualization

```python
# Let's see the current state of our dataset
sales_data.head()
```

```python
# Withdraw saleid as index by resetting
sales_data_viz = sales_data.reset_index()
sales_data_viz.head()
```

```python
# Let's assign MOS-UTC as index and sort
sales_data_viz = sales_data_viz.set_index('MOS-UTC').sort_index()
sales_data_viz.head()
```

```python
# Display the offset from UTC values along a timeline

sales_data_viz['OffsetUTC'].plot(figsize=(10, 6))

# The display wasn't informative
```

```python
# We'll display a scatter plot that shows the type of retailer shop selling company products in the different timezones around the globe

plt.scatter(sales_data_viz['OffsetUTC'], sales_data_viz['RetailerType'])
```

```python
# Inference:-
# Online Retailers are only successful in Europe (GMT and time zones 1 and 2 hours ahead)
# No Warehouse store or Mall are registered in The GMT. Same for Discount Retailer.
```

```python
# We want to create a bar chart to show how the sale amounts differ in the various time zones
# First, we group by timezone and sum all sales in them
sales_data_viz.groupby('TimeZone').sum()
```

```python
# Let's store this in a new variable
total_sales_revenue_by_time_zone = sales_data_viz.groupby('TimeZone').sum()
total_sales_revenue_by_time_zone
```

```python
# Plot the bar chart
plt.bar(total_sales_revenue_by_time_zone.index, 
        height = total_sales_revenue_by_time_zone['Sales Revenue ($)'])
```

```python
# Formatting the bar chart

# Set size of the figure
plt.figure(figsize = (7, 6))

# Specify bar colors
plt.bar(total_sales_revenue_by_time_zone.index, 
        height = total_sales_revenue_by_time_zone['Sales Revenue ($)'], 
        color = "#108A99")

# Adjust x and y ticks
plt.xticks(rotation = 45, fontsize = 13) 
plt.yticks(fontsize = 13)

# Add title and label
plt.title("Total Sales Revenue by Time Zone", fontsize = 18, fontweight = "bold") 
plt.ylabel("Sales Revenue ($)", fontsize = 13 ) 

plt.show()
```
